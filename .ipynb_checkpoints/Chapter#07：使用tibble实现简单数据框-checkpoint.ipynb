本章节使用的是数据框的一种变体：tibble。

```r
# 使用的核心R包
library(tidyverse)
```

R语言中的tibble和dataframe都是数据结构，用于存储和处理数据，但它们之间存在一些重要的区别。以下是tibble和dataframe之间的主要区别的表格对比：

| 特征                 | tibble                      | dataframe                  |
|----------------------|-----------------------------|----------------------------|
| 包含包              | 需要加载tidyverse包        | 基本R的一部分             |
| 列名处理            | 自动将非法列名修正为合法   | 不自动修正列名             |
| 数据显示            | 显示较少的数据             | 显示所有的数据             |
| 子集选择            | 不支持 "\$"运算符              | 支持"\$"运算符                |
| 行名处理            | 不允许行名为数字           | 允许行名为数字             |
| 数据框的性质        | 不会将字符向量转换为因子   | 默认将字符向量转换为因子   |
| 输出方式            | 更美观的输出方式           | 基本的文本输出方式         |
| 数据类型推断        | 更智能的数据类型推断       | 基于读入数据时的设置       |
| 列名和行名补全      | 自动补全列名和行名         | 不自动补全列名和行名         |
| 操作符重载          | 支持%>%操作符（管道操作符） | 不支持%>%操作符              |
| 默认的数据类型      | 使用dbl作为数值列的默认类型 | 使用numeric作为数值列的默认类型 |
| 数据切片和过滤      | 使用filter()和slice()函数 | 使用subset()函数和下标操作   |
| 数据添加和修改      | 使用mutate()函数            | 使用$操作符或赋值操作符       |
| 数据排序            | 使用arrange()函数           | 使用order()函数或sort()函数   |
| 缺失值处理          | 使用is.na()和complete()函数 | 使用is.na()和na.omit()函数     |

总的来说，tibble是dataframe的现代化替代品，它在数据处理和显示方面提供了更多的便利性和一致性，尤其是在使用tidyverse包进行数据分析时。然而，如果你需要与旧代码或基本R函数交互，你可能还需要使用dataframe。

其他的R包使用的是标准数据框，因此可以将数据框转换为tibble。

使用到的函数为：as_tibble()函数：

```r
as_tibble(iris)
```

可以通过tibble()函数使用一个向量来创建新tibble，tibble()函数会自动重复长度为1的输入，并且可以使用刚刚创建的新变量：

```r
tibble(
    x = 1:5,
    y = 1,
    z = x ^ 2 + y
)
```

在tibble中可以使用在R语言中无效的变量名称作为列名称：

```r
# 需要使用反引号
tb <- tibble(
    ':)' = "smile",
    ' ' = "space",
    '2000' = "number"
)

tb
```

如果要在ggplot2和dplyr等其他R包中使用这些变量，也需要使用反引号。

创建tibble的另一种方法是使用tribble()函数，tribble是transposed tibble的缩写。

tribble()函数是定制化的，可以对数据按照行进行编码：列标题由公式（以~开头）定义，数据条目以逗号分隔，这样就可以用易读的方式对少量数据进行布局：

```r
tribble(
    ~x, ~y, ~z,
    #--/--/----
    "a", 2, 3.6,
    "b", 1, 8.5
)
```

这里的 #--/--/--- 并不是tribble()语法，只是书写习惯，用来明确指出标题行的位置。

tibble和传统data.frame的使用方法主要有两处不同：打印和取子集。

tibble的打印方法进行了优化，只显示了前10行的结果，并且列也是适合屏幕的，这种方法是适合大数据集的。

```r
tibble(
    a = lubridate::now() + runif(1e3) * 86400,
    b = lubridate::today() + runif(1e3) * 30,
    c = 1:1e3,
    d = runif(1e3),
    e = sample(letters, 1e3, replace = TRUE)
)
```

在打印数据框的时候，tibble的这种设计避免了输出占满整个控制台。

但是有时候需要比默认显示更多的输出，这时候就需要设置几个选项。

```r
# 明确使用print()函数来打印数据框
install.packages("nycflights13")
library(nycflights13)
nycflights13::flights %>%
    print(n = 10, width = Inf)  # 这里使用了管道操作，n = 10意味着打印出来了10行，width = Inf则意味着将所有的列都打印了出来
```

还可以通过设置以下选项来控制默认的打印方式。

• options(tibble.print_max = n, tibble.pring_min = m)：如果多于 m 行，则只打印出 n行。options(tibble.print_min = Inf) 表示总是打印所有行。

• options(tibble.width = Inf) 表示总是打印所有列，不考虑屏幕的宽度。

```r
# 使用R语言的内置数据查看器
nycflights13::flights %>%
  View()
```

如果想要提取单个变量，那么就需要一些新工具，如 $  和 [[ 。

[[ 可以按照名称或者位置来提取变量；$只能够按名称提取变量，单可以减少一些输入：

```r
df <- tibble(
    x = runif(5),
    y = runif(5)
)
```

```r
df
```

```r
# 按名称提取
df$x
```

```r
# 按位置提取
df[[1]]
```

```r
# 使用管道进行操作，需要使用特殊占位符 .
df %>% .$x
```

与data.frame相比，tibble更加的严格：它不能够进行部分匹配，如果想要访问的列不存在，它会生成一条警告信息。

**练习**

```r
# 如何识别一个对象是否为 tibble
class(mtcars)
```

```r
is_tibble(mtcars)
```

```r
class(as_tibble(mtcars))
```

```r
is_tibble(as_tibble(mtcars))
```

```r
# 对比在 data.frame 和等价的 tibble 上进行的以下操作。
df <- data.frame(abc = 1, xyz = "a")
df$x
df[, "xyz"]
df[,c("abc", "xyz")]
```

```r
df <- as_tibble(df)
df$x
df[, "xyz"]
df[,c("abc", "xyz")]
```

$ 运算符将匹配以其后的名称开头的任何列名称。 由于存在名为 xyz 的列，因此表达式 df\$x 将扩展为 df\$xyz。 \$ 运算符的这种行为可以节省一些击键次数，但可能会导致意外使用与您想象的不同的列。

对于 data.frames，对于 [，返回的对象类型因列数而异。 如果是一列，则不会返回 data.frame，而是返回一个向量。 如果有多个列，那么它将返回一个 data.frame。 如果您知道要传递的内容，这很好，但假设您执行了 df[ , vars] ，其中 vars 是变量。 那么该代码的作用取决于长度（变量），并且您必须编写代码来解决这些情况或风险错误。

```r
# 如果将一个变量的名称保存在一个对象中，如 var <- "mpg"，如何从 tibble 中提取出这个变量？

# 创建一个示例tibble
my_tibble <- tibble(mpg = c(20, 30, 25), cyl = c(4, 6, 4))

# 将变量名称保存在一个对象中
var <- "mpg"

# 使用$运算符和反引号提取变量
variable <- my_tibble$`var`

# 打印提取的变量
print(variable)
```
