```r
# 手动调整颜色
ggplot(data = mpg) +
  geom_point(mapping = aes(x = displ, y = hwy), color = "red", alpha = 0.5)
```

```r
# "错误"代码
ggplot(data = mpg) + geom_point(mapping = aes(x = displ, y = hwy, color = "blue"))
```

上述代码的问题在于color="blue"在aes()中，这导致了在进行映射的时候，选择了"blue"变量来进行颜色的划分，但是问题在于mpg中并没有blue变量，所以带来的结果就是没有颜色的划分，而是采用了默认的情况。

将一个连续变量映射为 color、size 和 shape。对分类变量和连续变量来说，这些图形属性的表现有什么不同？

也会呈现出连续，比如说色彩从红到蓝的连续的变化；但是shape不能够被用于标注连续变量，因为shape是分类的，并不呈现出连续性。

```r
# 创建一个散点图，分别展示使用前和使用后的区别
p_with_legend <- ggplot(data = mpg, aes(x = cty, y = hwy, color = class)) +
  geom_point(show.legend = TRUE) +
  labs(title = "With Legend",
       x = "City MPG",
       y = "Highway MPG",
       color = "Car Class") +
  theme_minimal()

p_without_legend <- ggplot(data = mpg, aes(x = cty, y = hwy, color = class)) +
  geom_point(show.legend = FALSE) +
  labs(title = "Without Legend",
       x = "City MPG",
       y = "Highway MPG") +
  theme_minimal()

# 显示绘图
print(p_with_legend)
print(p_without_legend)
```

在第一个图中，我们设置了 geom_point(show.legend = TRUE)，因此图例会显示颜色（color）这一图层。而在第二个图中，我们设置了 geom_point(show.legend = FALSE)，因此图例中不会显示颜色这一图层。

```r
# 上面的代码存在着重复，可以将 mapping 映射放到 全局图层中
ggplot(data = mpg, mapping = aes(x = displ, y = hwy)) +
 geom_point() +
 geom_smooth()
```

如果将映射放在几何对象函数中，那么 ggplot2 会将其看作这个图层的局部映射，它将使用这些映射扩展或覆盖全局映射，但仅对该图层有效。这样一来，我们就可以在不同的图层中显示不同的图形属性：

```r
ggplot(data = mpg, mapping = aes(x = displ, y = hwy)) +
 geom_point(mapping = aes(color = class)) +
 geom_smooth()
```

```r
# fill，使用变量，自动叠块
ggplot(data = diamonds) +
 geom_bar(mapping = aes(x = cut, fill = clarity))
```

position = "identity" 将每个对象直接显示在图中。这种方式不太适合条形图，因为
条形会彼此重叠。为了让重叠部分能够显示出来，我们可以设置 alpha 参数为一个较小
的数，从而使得条形略微透明

```r
ggplot(
 data = diamonds,
 mapping = aes(x = cut, fill = clarity)
) +
 geom_bar(alpha = 1/5, position = "identity")
ggplot(
 data = diamonds,
 mapping = aes(x = cut, color = clarity)
) +
 geom_bar(fill = NA, position = "identity")
```

position = "fill" 的效果与堆叠相似，但每组堆叠条形具有同样的高度，因此这种条
形图可以非常轻松地比较各组间的比例：

```r
ggplot(data = diamonds) +
 geom_bar(
 mapping = aes(x = cut, fill = clarity),
 position = "fill"
 )
```

position = "dodge" 将每组中的条形依次并列放置，这样可以非常轻松地比较每个条形
表示的具体数值：

```r
ggplot(data = diamonds) +
 geom_bar(
 mapping = aes(x = cut, fill = clarity),
 position = "dodge"
 )
```

```r
# 位置调整
ggplot(data = mpg) +
 geom_point(
 mapping = aes(x = displ, y = hwy),
 )

 ggplot(data = mpg) +
 geom_point(
 mapping = aes(x = displ, y = hwy),
 position = "jitter"
 )
```

为 hwy 和 displ 的值都进行了舍入取整，所以这些点显示在一个网格上时，很多点彼此重叠了。这个问题称为过绘制。点的这种排列方式很难看出数据的聚集模式。


通过将位置调整方式设为“抖动”，可以避免这种网格化排列。position = "jitter" 为每个数据点添加一个很小的随机扰动，这样就可以将重叠的点分散开来，因为不可能有两个点会收到同样的随机扰动。

coord_flip() 函数可以交换 x 轴和 y 轴。当想要绘制水平箱线图时，这非常有用。

```r
ggplot(data = mpg, mapping = aes(x = class, y = hwy)) +
 geom_boxplot()

ggplot(data = mpg, mapping = aes(x = class, y = hwy)) +
 geom_boxplot() +
 coord_flip()
```

coord_polar() 函数使用极坐标系。

```r
bar <- ggplot(data = diamonds) +
 geom_bar(
 mapping = aes(x = cut, fill = cut),
 show.legend = FALSE,
 width = 1
 ) +
 theme(aspect.ratio = 1) +
 labs(x = NULL, y = NULL)
bar + coord_flip()
bar + coord_polar()
```
