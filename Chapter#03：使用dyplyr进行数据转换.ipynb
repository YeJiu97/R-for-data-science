列名下面有一行 3 个或 4 个字母的缩写。它们描述了每个变量的类型。
- int 表示整数型变量。
- dbl 表示双精度浮点数型变量，或称实数。
- chr 表示字符向量，或称字符串。
- dttm 表示日期时间（日期 + 时间）型变量。

另外还有三种类型：
- lgl 表示逻辑型变量，是一个仅包括 TRUE 和 FALSE 的向量。
- fctr 表示因子，R 用其来表示具有固定数目的值的分类变量。
- date 表示日期型变量。

dplyr的核心函数有5个：
1. filter()：按照值来进行观测
2. arrange()：对行进行重新排序
3. select()：按照名称选取变量
4. mutate()：使用现有的变量创建新变量
5. summarize()：将多个值总结为一个摘要统计

这些函数都可以和group_by()函数联合起来进行使用，group_by()可以修改以上每个函数的作用范围。

arrange()函数的工作方式与filter()函数非常的相似，但是arrange()函数的核心不是选择行而是改变行的顺序。

```r
# 语法
# arrange(dataset, col_01, col_02, col_03)
# 解释：将dataset的选中的col进行排序，先col_01，后col_02
arrange(flights, year, month, day)
```

```r
# 使用 desc() 可以进行降序排序
arrange(flights, desc(arr_delay))
```

```r
# select(dataset, col_01, col_02)
select(flights, year, month, day)
```

```r
select(flights, year:day)  # 从year到day之间的所有列，包括year和day
```

```r
select(flights, -(year:day))
```

还可以在 select () 函数中使用一些辅助函数。
- starts_with("abc")：匹配以“abc”开头的名称。
- ends_with("xyz")：匹配以“xyz”结尾的名称。
- contains("ijk")：匹配包含“ijk”的名称。
- num_range("x", 1:3)：匹配 x1、x2 和 x3。

select() 可以重命名变量，但我们很少这样使用它，因为这样会丢掉所有未明确提及的变
量。我们应该使用 select() 函数的变体 rename() 函数来重命名变量，以保留所有未明确
提及的变量

```r
rename(flights, tail_num = tailnum)
```

mutate()可以用来添加新变量。

```r
flights_sml <- select(flights,
    year:day,
    ends_with("delay"),
    distance,
    air_time
)

flights_sml
```

```r
mutate(flights_sml,
    gain = arr_delay - dep_delay,
    speed = distance / air_time * 60
)
```

```r
transmute(flights,
 gain = arr_delay - dep_delay,
 hours = air_time / 60,
 gain_per_hour = gain / hours
)
```

```r
temperature_data <- data.frame(
    city = c("New York", "New York", "Los Angeles", "Los Angeles", "Chicago", "Chicago"),
    date = c("2023-08-01", "2023-08-02", "2023-08-01", "2023-08-02", "2023-08-01", "2023-08-02"),
    temperature = c(80, 85, 75, 78, 70, 72)
)

average_temperatures <- temperature_data %>%
    group_by(city) %>%
    summarize(avg_temperature = mean(temperature))
```

```r
by_dest <- group_by(flights, dest)
delay <- summarize(by_dest,
    count = n(),
    dist = mean(distance, na.rm = TRUE),
    delay = mean(arr_delay, na.rm = TRUE)
)
delay <- filter(delay, count > 20, dest != "HNL")

ggplot(data = delay, mapping = aes(x = dist, y = delay)) +
    geom_point(aes(size = count), alpha = 1/3) +
    geom_smooth(se = FALSE)
```

```r
# 创建示例数据集
temperature_data <- data.frame(
  city = c("New York", "New York", "Los Angeles", "Los Angeles", "Chicago", "Chicago"),
  temperature = c(80, 85, 75, 78, 70, 72)
)

# 对每个城市的温度数据进行汇总
summary_data <- temperature_data %>%
  group_by(city) %>%
  summarize(
    avg_temperature = mean(temperature),
    max_temperature = max(temperature)
  )

print(summary_data)
```
