# 简介

建立模型的目的是提供一个简单的、低维度的数据集摘要。

使用模型的目的是将数据划分为：

1. 模式：在数据科学和统计学中，模型用于通过一组输入预测输出
2. 残差：残差则是在统计模型中，真实观测值与模型预测值之间的差异

建模可以分为两个阶段：

1. 定义一个模式族来表示一种精确但是一般性地模式，这种模式是我们想要捕获地
2. 生成一个拟合模型，方法是从模型族中找出最接近数据地一个模型

拟合模型只是模型族中与数据最接近地一个模型，这并不意味着这个模型是真的。

**模型的目标不是发现真理，而是获得简单但有价值的近似。**

# 一个简单模型

使用sim1数据集，包含两个连续型变量x和y，绘制结果如下所示：

```r
ggplot(sim1, aes(x, y)) +
    geom_point()
```

通过可视化，可以发现一个非常强地模式，可以认为x和y之间地关系是线性的。

```r
# 这部分代码使用 tibble 函数创建了一个数据框 models
# 这个数据框包含两列 a1 和 a2，每列有250个随机数。

models <- tibble(
    a1 = runif(250, -20, 40),  # a1 列生成的是从-20到40之间的均匀分布的随机数
    a2 = runif(250, -5, 5)  # a2 列生成的是从-5到5之间的均匀分布的随机数
)
```

```r
# 绘制图形
ggplot(sim1, aes(x, y)) +  # 其中 sim1 应该是一个包含 x 和 y 的数据框
    geom_abline(  # geom_abline 函数添加了多条线，每条线的截距和斜率由 models
        aes(intercept = a1, slope = a2),
        data = models, alpha = 1/4  # 透明度为25%
    ) +
    geom_point()
```

这张图中有 250 个模型，但是通过观察可以看出，大部分模型都是非常糟糕的。

需要一种方法来量化数据与模型之间的差距，然后找出最佳的a_0和a_1的值。

如上图所示，直线是模型的可视化结果，而数据点和模型的垂直距离则是预测与真实值的差距。

```r
# 将模型转换为R函数
model1 <- function(a, data){
    a[1] + data$x * a[2]  # 其中 a[1] 用作线性模型的截距，a[2] 用作斜率
    }
```

```r
model1(c(7, 1.5), sim1)  
# 即对数据框 sim1 中的每一行应用公式 y=7+1.5×x
```

我们可以通过上述的代码计算出来我们对数据集中的每个数据的预测结果，基于我们定义的模型，接下来要做的是计算每个点的差值的和。

```r
# 均方根误差
measure_distance <- function(mod, data){
    diff <- data$y - model1(mod, data)
    sqrt(mean(diff ^ 2))
}
```

```r
measure_distance(c(7, 1.5), sim1)
```

接着可以使用purrr来计算前面定义的所有模型和数据见间的距离了：

```r
sim1_dist <- function(a1, a2) {
    measure_distance(c(a1, a2), sim1)
}
```

```r
models <- models %>% mutate(dist = purrr::map2_dbl(a1, a2, sim1_dist))
```

```r
models
```

将最好的 10 个模型覆盖到数据上：

```r
ggplot(sim1, aes(x, y)) +
    geom_point(size = 2, color = "grey30") +
    geom_abline(
        aes(intercept = a1, slope = a2, color = -dist),
        data = filter(models, rank(dist) <= 10)
)
```

我们将这张图和原先的图进行比较，可以发现保留下来的模型基本上都在数据点的分布范围内了。

我们还可以将这些模型看作观测，并使用由 a1 和 a2 组成的一张散点图来表示它们，还是使用 -dist 进行上色。虽然这样不能直接看到模型与数据间的比较，但我们可以同时看到很多模型。

```r
ggplot(models, aes(a1, a2)) + geom_point(
    data = filter(models, rank(dist) <= 10), size = 4, color = "red" ) +  # 排名前10的点，用红色标出这些点的位置
geom_point(aes(colour = -dist))
```

相较于检查多个随机模型，我们使用一种更加系统化的方法来找出模型参数（这种方法称为网格搜索法）。

首先，我们生成一张分布均匀的数据点网格，然后将这个网格与前面图 中的 10 个最佳模型绘制在一张图中，凭借最佳模型在网格中的位置就可以找出模型参数的粗略值：

```r
grid <- expand.grid(
    a1 = seq(-5, 20, length = 25),
    a2 = seq(1, 3, length = 25)
) %>%
mutate(dist = purrr::map2_dbl(a1, a2, sim1_dist))
```

```r
grid %>% ggplot(aes(a1, a2)) + geom_point(
    data = filter(grid, rank(dist) <= 10),
    size = 4, colour = "red"
) + geom_point(aes(color = -dist))
```

如果将这 10 个最佳模型重新覆盖到原始数据上，就可以看出效果还是很不错的：

```r
ggplot(sim1, aes(x, y)) + geom_point(size = 2, color = "grey30") + geom_abline(
    aes(intercept = a1, slope = a2, color = -dist),
    data = filter(grid, rank(dist) <= 10)
)
```

可以设想不断细化网格来最终找出最佳模型。

有一个更好的方法可以解决这个问题，这种方法是名为“牛顿—拉夫逊搜索”的数值最小化工具。

先选择一个起点，环顾四周找到最陡的斜坡，并沿着这个斜坡向下滑行一小段，然后不断重复这个过程，直到不能再下滑为止。

```r
# 使用 optim() 函数
best <- optim(c(0, 0), measure_distance, data = sim1)
```

```r
best$par
```

```r
ggplot(sim1, aes(x, y)) +
    geom_point(size = 2, color = 'grey30') +
    geom_abline(intercept = best$par[1], slope = best$par[2])
```

无须过多担心 optim() 函数的工作细节。现在重要的是要建立直觉。

如果具有定义模型与数据集间距离的函数，以及可以通过修改模型参数使距离最小化的算法，那么我们就可以找出最佳模型。

以上这种方法的好处是，只要能够写出方程来表示模型族，那么你就可以使用这种方法。

对于这个模型，我们还可以使用另一种方法，因为它是一个更广泛模型族的一种特殊情况，即是线性模型。

线性模型的一般形式是 y = a_1 + a_2 * x_1 + a_3 * x_2 + … + a_n * x_(n - 1)。

```r
# 使用lm()函数来进行实现
sim1_mod <- lm(y ~ x, data = sim1)  # 进行了拟合
```

```r
coef(sim1_mod)
```

这与我们使用 optim() 函数得到的结果完全一致。lm() 函数使用的并不是 optim()，而是利用了线性模型的数学结构。实际上，lm() 使用了一种非常复杂的算法，通过几何学、微积分和线性代数间的一些关系，它只需要一个步骤就可以找出最近似的模型。这种方法的速度非常快，而且一定能找到全局最小值。

# 模型可视化

找出模型未捕获的信息也是非常有用的，即所谓的残差，它是数据去除预测值后剩余的部分。残差是非常强大的，因为它允许我们使用模型去除数据中显著的模式，以便对剩余的微妙趋势进行研究。

要想对模型的预测进行可视化表示，首先要生成一个分布均匀的数值网格，以覆盖数据所在区域。完成这个任务最简单的方式就是使用 modelr::data_grid() 函数，其第一个参数是一个数据框，对于随后的每个参数，它都会找出其中的唯一值，然后生成所有组合：

```r
# 预测
grid <- sim1 %>% data_grid(x)
grid
```

接着添加预测值。我们使用的是 modelr::add_predictions() 函数，其参数是一个数据框和一个模型。

```r
grid <- grid %>% add_predictions(sim1_mod)
grid
```

```r
# 下一步是绘制预测值。

ggplot(sim1, aes(x)) + geom_point(aes(y = y)) +
geom_line( aes(y = pred), data = grid, colour = "red", size = 1)
```

与预测值相对的是残差。预测值可以告诉我们模型捕获的模式，残差则表示模型漏掉的部分。残差就是我们前面计算过的观测值与预测值间的距离。

我们可以使用 add_residuals() 函数将残差添加到数据中，这个函数与 add_predictions()非常相似。

```r
sim1 <- sim1 %>% add_residuals(sim1_mod)
```

```r
sim1
```

对于残差可以反映出模型的哪些信息，有几种不同的理解方法。其中一种方法是简单地绘制频率多边形图，以帮助我们理解残差的分布

```r
ggplot(sim1, aes(resid)) + geom_freqpoly(binwidth = 0.5)
```

这种方法可以反映出模型的质量：模型预测值与实际观测值的差别有多大？注意，残差的平均值总是为 0。

这句话在特定情况下是正确的，尤其是在使用最小二乘法拟合线性回归模型时。最小二乘法的目标是最小化残差的平方和，即 $ \sum (y_i - \hat{y}_i)^2 $ ，其中 $ y_i $ 是观测值，$ \hat{y}_i $ 是模型预测值。在这种情况下，残差的平均值为零的原因可以从数学和统计的角度来解释：

在最小二乘线性回归中，残差 $ e_i = y_i - \hat{y}_i $ 。如果我们将所有的残差相加，即 $ \sum e_i $ ，这等同于 $ \sum (y_i - \hat{y}_i) $ 。由于线性回归模型包括一个常数项（截距），该模型会自动调整截距的值，使得残差之和为零。换句话说，线性回归模型会通过调整截距来“平衡”预测值和实际值，从而使残差的总和（因此平均值）等于零。

从统计的角度看，残差的平均值为零是模型对数据拟合的一个结果。在最小二乘回归中，通过选择斜率和截距，使得 $ \sum (y_i - \hat{y}_i) = 0 $ 。这表明模型没有系统偏误，或者说模型没有未解释的系统趋势。这也是为什么在回归分析中常常检查残差图：如果残差的平均值不为零，可能表示模型存在规范错误或遗漏了某些重要的解释变量。

尽管在最小二乘回归中残差的平均值通常为零，但在其他一些模型或方法中，这个结论可能不成立。例如，在不包括截距的模型中，或者在使用其他优化准则（如最小绝对偏差回归）时，残差的平均值可能不为零。

因此，对于最小二乘线性回归模型而言，残差的平均值总是为零是一个正确的说法。这反映了模型的一个基本属性，即在没有模型偏误的情况下，预测值和实际值在统计上是平衡的。

我们经常会使用残差代替原来的预测变量来重新绘图，这主要是为了评估模型的拟合质量，识别可能的数据问题，以及验证模型假设。

检查模型假设：在线性回归模型中，基本假设包括

- 残差的独立性：观测之间的残差应相互独立
- 残差的正态性：残差应呈正态分布
- 残差的方差齐性（同方差性）：对于所有的预测值，残差的方差应该是常数

通过绘制残差图（例如，残差与拟合值的散点图），我们可以直观地检查这些假设是否得到满足。

残差图可以帮助识别数据中的异常值和杠杆点。异常值是那些与其他数据点明显不同的观测值，它们的残差往往非常大。杠杆点是那些具有非常不寻常的预测变量值的观测点，它们即使残差不大，也可能对回归结果产生不成比例的影响。

通过分析残差图，可以发现需要对模型进行改进的地方。例如，如果残差与预测值的图显示出非线性模式，这可能表明需要在模型中添加更高阶的项（如二次项或交互项）。

```r
ggplot(sim1, aes(x, resid)) + geom_ref_line(h = 0) + geom_point()
```

由上图可知，残差应该是随机的噪声，这表明我们的模型非常好地捕获了数据集中的模式。

# 公式和模型族

如果想看 R 到底进行了什么转换，可以使用 model_matrix() 函数。这个函数接受一个数据框和一个公式，并返回一个定义了模型方程的 tibble，其中每一列都关联到方程的一个系数，方程的形式总是类似于 y = a_1 * out1 + a_2 * out_2。对于最简单的情况，y ~ x1，这个函数会返回以下有趣的结果：

```r
df <- tribble(
    ~y, ~x1, ~x2,
    4, 2, 5,
    5, 1, 6
)
```

```r
model_matrix(df, y ~ x1)
```

R 向模型加入截距项的方法是，加入一个值全是 1 的列。默认情况下，R 总是加入这一列。如果不想要截距项，那么你必须使用 -1 来明确丢弃它

```r
model_matrix(df, y ~ x1 - 1)
```

如果向模型中添加更多变量，那么模型矩阵当然也会随之增长：

```r
model_matrix(df, y ~ x1 + x2)
```

如果预测变量是连续的，那么从公式转换为方程是很简单的；但当预测变量是分类变量时，事情就有点复杂了。假设存在公式 y ~ sex，其中 sex 的值要么是男性要么是女性，那么将其转换成 y = x_0 + x_1 * sex 就没有意义了，因为 sex 不是数值，我们不能对它使用乘法。相反，R 的做法是将其转换为 y = x_0 + x_1 * sex_male，如果 sex 为男性，那么 sex_male 的值就是 1，否则其值就是 0：

```r
df <- tribble(
    ~ sex, ~ response,
    "male", 1,
    "female", 2,
    "male", 1
)
```

```r
model_matrix(df, response ~ sex)
```

你可能想知道，为什么 R 没有同时建立一个 sexfemale 列。问题是，这样创建出来的列是可以基于其他列完美预测的（即 sexfemale = 1 - sexmale）。

多重共线问题，简单来讲就是这样做会建立一个过于灵活的模型族，从而生成对数据拟合效果相同的无数个模型。

在模型训练过程中，某个分类变量（因子变量）被设置了特定的水平（类别）。当使用该模型进行预测时，如果输入数据中含有在训练集中未出现过的新水平，R不知道如何处理这些新水平，因此会抛出错误。

在你的案例中，可能存在一个因子变量 x。在模型训练（使用 lm(), glm() 等）时，x 包含了一些类别，例如 a, b, c 等。但是在你尝试进行预测的新数据中，x 出现了一个新的水平 e，这个水平在模型训练时的数据中没有出现过。

移除或替换新水平：在对新数据进行预测前，检查并移除或替换那些在训练数据中未出现过的水平。这可以通过设定默认类别或将其归为其他已知类别来实现。

扩展因子水平：如果新的水平可能具有实际意义，且应包括在预测中，可以在模型训练阶段确保所有潜在的水平都被包括为因子的水平。

重新训练模型：如果新的水平确实重要，考虑使用包含所有相关水平的数据集重新训练模型。

$$
y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \epsilon
$$

每个自变量对因变量 y 的影响是独立的，即 x1 对 y 的影响不依赖于 x2 的值，反之亦然。

```r
mod2 <- lm(y ~ x1 * x2, data = sim3)
```

这个模型不仅包含 x1 和 x2 对 y 的独立影响，还包括了它们的交互作用。这意味着 x1 和 x2 的组合效果对 y 的影响也被模型所考虑。

$$
y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \beta_3 x_1 x_2 + \epsilon
$$

如果使用 + 添加变量，那么模型会独立地估计每个变量的效果，不考虑其他变量。如果使用*，那么拟合的就是所谓的交互项。

接下来我们试着对模型进行可视化。因为有两个连续型预测变量，所以我们可以将模型想象为一个三维表面。

```r
ggplot(grid, aes(x1, x2)) + geom_tile(aes(fill = pred)) + facet_wrap(~ model)
```

观察这两张图，我们看不出这两个模型有什么明显区别。但这是一种错觉，因为我们的眼
睛和大脑不擅长精确分辨颜色的深浅。

```r
ggplot(grid, aes(x1, pred, color = x2, group = x2)) + 
geom_line() + facet_wrap(~ model)
```

```r
ggplot(grid, aes(x2, pred, color = x1, group = x1)) +
geom_line() +
facet_wrap(~ model)
```

这说明两个连续型变量的交互项的作用方式与一个分类变量和一个连续变量的交互项基本
相同。交互项说明了两个变量是相互影响的，如果要预测 y 值，那么必须同时考虑 x1 的
值和 x2 的值。

还可以在模型公式中进行变量转换。例如，$log(y) ~ sqrt(x1) + x2$ 可以转换为 $log(y) = a_1 + a_2 * sqrt(x1) + a_3 * x2$。如果想要使用 +、*、^ 或 - 进行变量转换，那么就应该使用 I() 对其进行包装，以便 R 在处理时不将它当作模型定义的一部分。

例如，$y ~ x + I(x ^ 2)$ 会转换为 $y = a_1 + a_2 * x + a_3 * x^2$。如果忘记使用 I()，将公式写成 $y ~ x ^ 2 + x$，那么 R 就会计算 $y ~ x * x + x。x * x$ 表示 x 和自己的交互项，

```r
df <- tribble(
    ~y, ~x,
    1, 1,
    2, 2,
    3, 3
)
```

```r
model_matrix(df, y ~ x^2 + x)
```

```r
model_matrix(df, y ~ I(x^2) + x)
```

变量转换非常有用，因为你可以使用它们来近似表示非线性函数。

```r
model_matrix(df, y ~ poly(x, 2))
```

但使用 poly() 函数时有一个很大的问题：多项式的值会超出数据范围，很容易接近正无穷或负无穷。更安全的一种方式是使用自然样条法 splines::ns()：

```r
library(splines)
model_matrix(df, y ~ ns(x, 2))
```

```r
sim5 <- tibble(
    x = seq(0, 3.5 * pi, length = 50),
    y = 4 * sin(x) + rnorm(length(x))
)
```

```r
ggplot(sim5, aes(x, y)) + geom_point()
```

我们使用这份数据来拟合 5 个模型：

```r
mod1 <- lm(y ~ ns(x, 1), data = sim5)
mod2 <- lm(y ~ ns(x, 2), data = sim5)
mod3 <- lm(y ~ ns(x, 3), data = sim5)
mod4 <- lm(y ~ ns(x, 4), data = sim5)
mod5 <- lm(y ~ ns(x, 5), data = sim5)
```

```r
grid <- sim5 %>% 
data_grid(x = seq_range(x, n = 50, expand = 0.1)) %>%
gather_predictions(mod1, mod2, mod3, mod4, mod5, .pred = "y")
```

```r
ggplot(sim5, aes(x, y)) +
geom_point() +
geom_line(data = grid, color = "red") +
facet_wrap(~ model)
```

注意，当使用模型在数据范围外进行推断时，效果明显非常差。这是使用多项式近似函数的一个缺点。但这是所有模型都具有的一个实际问题：当对未知数据进行外推时，模型无法确保结果的真实性。你必须依靠相关的科学理论。

# 缺失值

对于变量间的关系，缺失值显然不能传达出任何信息，因此建模函数会丢弃包含缺失值的所有行。默认情况下，R 会不声不响地丢弃这种数据，

```r
#  options(na.action = na.warn)（我们在准备工作中运行过这行代码）可以确保我们收到一条警告信息
df <- tribble(
    ~x, ~y,
    1, 2.2,
    2, NA,
    3, 3.5,
    4, 8.3,
    NA, 10
)
mod <- lm(y ~ x, data = df)
```

要想阻止错误信息，可以设置 na.action = na.exclude：

```r
mod <- lm(y ~ x, data = df, na.action = na.exclude)
```

通过 nobs() 函数，你可以知道模型实际使用了多少个观测：

```r
nobs(mod)
```
